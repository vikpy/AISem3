# Q1. Read any csv file or any json file in df with 6 cols atleast and 1000 rows

# Q2. Split the file into test and train using pandas with a user input for the ratio of split

```
train_ratio = float(input("Give the train ratio value:"))
train_value = int(train_ratio*df.shape[0])
train = df.iloc[:train_value,:]
test = df.iloc[train_value:,:]
```

#Q4. Write a function to query if a values exists in the dataframe and return its complete column and complete row separately

```
def queryDf(df, value):
  row_colum = set()
  for column in df.columns.values:
    for row in df[column].index:
      if df.loc[row,column] == value :
        row_colum.add((row, column))
  return row_colum

queryDf(df, 34.19)
```

# Q6.From the list of columns , construct all combinations of 3 values and for each combination combine their row values as  a combined column

```
combinations = set()
for value1 in df.columns.values:
  for value2 in df.columns.values:
    for value3 in df.columns.values:
      if value1 != value2 and value1 != value3 and value2 != value3:
        combinations.add((value1, value2, value3))

combinations
```

```
for values in combinations:
  print('*'*100)
  print(values)
  print(np.array(df[list(values)]))
```
