```
# First Implementation Fully Connected Graph.
class FullyConnectedGraph:
  def __init__(self, *nodes):
    self._graph_nodes = nodes
    self._graph = None 
    import networkx as nx
    import matplotlib.pyplot as plt
    h=nx.complete_graph(len(nodes))
    nx.draw(h)
    plt.show()
    

  def implement_graph(self, type=5):
    '''
     Type is one of the 5 types discussed during the lecture.
    '''
    if type == 1:
      self._type1_implemetation()
    elif type == 2:
      self._type2_implemetation()
    elif type == 3:
      self._type3_implemetation()
    elif type == 4:
      self._type4_implemetation()
    elif type == 5:
      self._type5_implemetation()

  def _type1_implemetation(self):
    main_array = []
    for main_node in self._graph_nodes :
      intermediate_array = [main_node]
      for sub_node in self._graph_nodes: 
        if main_node != sub_node:
          intermediate_array.append(sub_node)
      main_array.append(intermediate_array)
    self._graph = main_array

  def _type2_implemetation(self):
    main_array = []
    for main_node in self._graph_nodes :
      intermediate_array = [main_node]
      inter_temp_array  = []
      for sub_node in self._graph_nodes: 
        if main_node != sub_node:
          inter_temp_array.append(sub_node)
      intermediate_array.append(inter_temp_array)
      main_array.append(intermediate_array)
    self._graph = main_array
  
  def _type3_implemetation(self):
    main_array = []
    for main_node in self._graph_nodes :
      intermediate_array = [main_node]
      inter_temp_array  = []
      for sub_node in self._graph_nodes: 
        if main_node != sub_node:
          inter_temp_array.append(sub_node)
      intermediate_array.append( tuple(inter_temp_array) )
      main_array.append(intermediate_array)
    self._graph = main_array

  def _type4_implemetation(self):
    main_array = []
    for main_node in self._graph_nodes :
      intermediate_array = [main_node]
      inter_temp_array  = []
      for sub_node in self._graph_nodes: 
        if main_node != sub_node:
          inter_temp_array.append(sub_node)
      intermediate_array.append(inter_temp_array)
      main_array.append(intermediate_array)
    self._graph = tuple(main_array ) 

  def _type5_implemetation(self):
    main_dictionary = {}
    for main_node in self._graph_nodes :
      intermediate_array = []
      for sub_node in self._graph_nodes: 
        if main_node != sub_node:
          intermediate_array.append(sub_node)
      main_dictionary[main_node] = intermediate_array
    self._graph = main_dictionary

  def __str__(self):
    return str(self._graph)
  
  def get_graph(self):
    return self._graph

```

```
fcg = FullyConnectedGraph('A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J')
for type_of_graph in range(1, 6):
  fcg.implement_graph(type_of_graph)
  print('-'*100)
  print(f'Type {type_of_graph} of Graph')
  print(fcg)
```
